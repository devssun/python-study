```python
# setting
from bs4 import BeautifulSoup
from selenium import webdriver
import time

def open_chrome_driver():
    path = 'chromedriver'
    driver = webdriver.Chrome(path)
    # 24 시간 운영하는 매장만 검색합니다
    driver.get('http://www.hollys.co.kr/store/korea/korStore.do?pageNo=1&sido=&gugun=&store=&allday=1')
    return driver

# 현재 보고있는 페이지에서 html parser 실행하기
def get_soup():
    source = driver.page_source
    soup = BeautifulSoup(source, 'html.parser')
    return soup

# table 코드 파싱하기
def parsing_cafe_list(soup):
    result = []
    tags = soup.select('td > a')
    for tag in tags:
        result.append(tag.text)
    return result

# 카페 이름과 주소 가져오기 
def get_store_info(list):
    titles = []
    address = []
    for i in range(0, len(list)):
        if i % 2 == 0:
            titles.append(list[i])
        else:
            address.append(list[i])
    return (titles, address)

# 페이지 이동
def move_page(page_no):
    element = driver.find_element_by_class_name("paging")
    element.find_element_by_link_text(str(page_no)).click()
```

```python
page_no = 1
driver = open_chrome_driver()
store_titles = []
store_address = []

for i in range(0, 5):
    soup = get_soup()
    cafe_list = parsing_cafe_list(soup)
    (titles, address) = get_store_info(cafe_list)
    store_titles = store_titles + titles
    store_address = store_address + address
    time.sleep(2)
    
    page_no = page_no + 1
    move_page(page_no)
```
